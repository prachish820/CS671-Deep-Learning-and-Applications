## Group 13: ai, bA, chA, lA, tA

```python
# importing libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data

root_dir = ".\Handwriting_Data"
class_dict = {"ai":0,"bA":1,"chA":2,"lA":3,"tA":4}

train_paths, test_paths = [],[]
for i in list(class_dict.keys()):
    class_dir = os.path.join(root_dir,i)
    train_dir = os.path.join(class_dir,"train")
    test_dir = os.path.join(class_dir,"dev")
```

```python
def read_data(path):
    
    train_dir = os.path.join(path,"train")
    test_dir = os.path.join(path,"dev")
    
    train_df, test_df = pd.DataFrame(),pd.DataFrame()
    for file in os.listdir(train_dir):
        file_path = os.path.join(train_dir,file)
        train_df = train_df.append(pd.read_csv(file_path, sep = " ", header = None), ignore_index = True)
    train_df.drop(0, inplace = True, axis = 1)
    
    for file in os.listdir(test_dir):
        file_path = os.path.join(test_dir,file)
        test_df = test_df.append(pd.read_csv(file_path, sep = " ", header = None), ignore_index = True)
    test_df.drop(0, inplace = True, axis = 1)
    
    return train_df, test_df
```

```python
class_dirs = []
for i in list(class_dict.keys()):
    class_dirs.append(os.path.join(root_dir,i))
```

```python
train_ai, test_ai = read_data(class_dirs[0])
train_bA, test_bA = read_data(class_dirs[1])
train_chA, test_chA = read_data(class_dirs[2])
train_lA, test_lA = read_data(class_dirs[3])
train_tA, test_tA = read_data(class_dirs[4])
```

```python
train_ai
```

```python
test_ai
```

```python
def normalize_data(data):
    m,n=data.shape
    for i in range(m):
        idx = data.iloc[i].isna().idxmax()
        x,y = [],[]
        for j in range(1,idx,2):
            x.append(data.iloc[i,j])
            y.append(data.iloc[i,j+1])
        for j in range(1,idx,2):
            data.iloc[i,j] = data.iloc[i,j]/max(x)
            data.iloc[i,j+1] = data.iloc[i,j+1]/max(y)
    return data
```

```python
train_ai, test_ai = normalize_data(train_ai), normalize_data(test_ai)
train_bA, test_bA = normalize_data(train_bA), normalize_data(test_bA)
train_chA, test_chA = normalize_data(train_chA), normalize_data(test_chA)
train_lA, test_lA = normalize_data(train_lA), normalize_data(test_lA)
train_tA, test_tA = normalize_data(train_tA), normalize_data(test_tA)
```

```python
train_dfs = [train_ai, train_bA, train_chA, train_lA, train_tA]
test_dfs = [test_ai, test_bA, test_chA, test_lA, test_tA]
```

```python
def gen_char(dfs):
    for i,df in enumerate(dfs):
        data = df.iloc[0]
        n = len(data)
        x,y = [],[]
        idx = data.isna().idxmax()
        for j in range(1,idx,2):
            x.append(data[j])
            y.append(data[j+1])
        plt.plot(x,y)
        plt.title(list(class_dict.keys())[i])
        plt.show()
```

```python
gen_char(train_dfs)
```

```python
# adding a label column to all dfs

for i,df in enumerate(train_dfs):
    df.insert(0,"label",i)
for i,df in enumerate(test_dfs):
    df.insert(0,"label",i)
```

```python
train_ai
```

```python
concat_dfs = pd.concat(train_dfs, ignore_index=False)           # concatenating dfs of all classes to form input df
concat_dfs.drop(concat_dfs.columns[-1],axis=1,inplace=True)     
concat_dfs = concat_dfs.fillna(0)                               # filling nans with 0
```

```python
concat_dfs.shape
```

# Building model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Input, Dense, SimpleRNN, LSTM, RNN, Masking, Dropout
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras import initializers
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
from keras.utils import pad_sequences
```

```python
# prepaparing input data
x = concat_dfs.iloc[:,1:].values
y = concat_dfs.iloc[:,0].values

max_len = int(x.shape[1]/2)
n_samples, n_timesteps = x.shape[0],max_len
n_features = 2
x = x.reshape((n_samples,n_timesteps,n_features))
```

```python
x.shape
```

```python
# splitting training data into train-validation data
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# prepare test data

test_concat = pd.concat(test_dfs, ignore_index=False)
test_concat.drop(test_concat.columns[-1],axis=1,inplace=True)
test_concat = concat_dfs.fillna(0)

x_test = test_concat.iloc[:,1:].values
y_test = test_concat.iloc[:,0].values

max_test_len = int(x_test.shape[1]/2)
n_test_samples, n_test_timesteps = x_test.shape[0],max_test_len
n_features = 2
x_test=x_test.reshape((n_test_samples, n_test_timesteps,n_features))
```

## Single recurrent layer model

```python
input_shape = (None, 2)
input_layer = Input(shape = input_shape)
masking_layer = Masking(mask_value=0.0, input_shape=input_shape)(input_layer)
rnn_layer = SimpleRNN(units=100, activation="tanh")(masking_layer)
#dropout_layer = Dropout(0.2)(rnn_layer)
output_layer = Dense(units=5, activation="softmax")(rnn_layer)

model1 = Model(inputs=input_layer, outputs=output_layer)
```

```python
model1.summary()
```

```python
my_callbacks = EarlyStopping(monitor='loss',min_delta=0.0001,patience=5,verbose=1)

model1.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
train_model1 = model1.fit(x_train,y_train,batch_size=32,epochs=1000,callbacks=my_callbacks,validation_data=(x_val,y_val),shuffle=True)
```

```python
history1 = train_model1.history

import pickle
with open('./models/RNN1_history.pkl','wb') as f:
    pickle.dump(history1,f)

# plot loss function
plt.plot(history1['loss'])
plt.ylabel('Average Error',fontsize=12)
plt.xlabel('Epoch',fontsize=12)
plt.savefig('./models/RNN1_loss.png')
plt.show()
```

```python
# saving model
model1.save('./models/RNN1_model.h5')
```

```python
# evaluate model on train, validation and test data
train_loss, train_acc = model1.evaluate(x_train,y_train)
val_loss, val_acc = model1.evaluate(x_val, y_val)
test_loss, test_acc = model1.evaluate(x_test, y_test)
loss, acc = model1.evaluate(x,y)
```

```python
# predicting train, validation and test data labels
train_out = model1.predict(x_train)
val_out = model1.predict(x_val)
test_out = model1.predict(x_test)
y_out = model1.predict(x)
```

```python
# confusion matrix on train, validation and test data
from tensorflow.math import confusion_matrix

def conf_mat(true,predicted):
    predicted_label = np.argmax(predicted,axis=1)
    mat = confusion_matrix(true,predicted_label)
    
    return mat
```

```python
print(conf_mat(y_train,train_out))
print(conf_mat(y_val,val_out))
print(conf_mat(y_test,test_out))
print(conf_mat(y,y_out))
```
